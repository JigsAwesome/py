```python
import matplotlib.pyplot as plt
expense_value=[2000,500,100,50]
items=['home rent','recharge','food','moj']
plt.pie(expense_value,labels=items, autopct='%0.1f%%', radius=2, explode=[0.1,0.2,0.1,0.2],shadow='true',startangle=180
       )
plt.show()
    
```
